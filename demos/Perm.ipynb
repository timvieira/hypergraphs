```python
from semirings import LazySort
from semirings.symbolic import Symbolic, simplify

def parse_perm(xs, Weight):
    n = len(xs)
    c = Weight.chart()
    for i in range(n):
        c[i,i+1] = Weight.lift(1, xs[i])
    for w in range(2, n+1):
        for i in range(n-w+1):
            k = i + w
            for j in range(i+1, k):
                c[i,k] += c[i,j] * c[j,k]
                c[i,k] += c[j,k] * c[i,j]
    return c[0,n]
```

```python
from semirings import flatten

elems = [x.data for x in parse_perm([1,2,3,4], LazySort)]

S = set()
n = 0
for z in sorted(elems, key=lambda x: tuple(flatten(x))):
    y = tuple(flatten(z))

    print(str(z), '*' if y in S else '')
    #assert y not in S
    S.add(y)
    n += 1
print('total:', n, 'unique:', len(S), 'max:', 4*3*2*1)
```
